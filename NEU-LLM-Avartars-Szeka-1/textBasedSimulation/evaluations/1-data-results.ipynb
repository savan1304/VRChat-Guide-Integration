```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Parameters (m, n)': ['(3, 3)', '(3, 5)', '(5, 3)', '(5, 5)', '(5, 7)', '(7, 5)', '(7, 7)'],
    'mAP Accross Models for Test Run 1': [2.785316971, 2.88120517, 2.341611295, 2.93594704, 2.769488107, 2.58801233, 2.589405872],
    'mAP Accross Models for Test Run 2': [2.831115471, 2.839172143, 2.767282138, 2.701529528, 2.816452691, 2.689142178, 2.649476326]
}

df = pd.DataFrame(data)

df['Average'] = df[['mAP Accross Models for Test Run 1', 'mAP Accross Models for Test Run 2']].mean(axis=1)

df = df.drop(columns=['mAP Accross Models for Test Run 1', 'mAP Accross Models for Test Run 2'])
df
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Parameters (m, n)'], df['Average'], marker='o', label='Average mAP')
plt.xlabel('Parameters (m, n)')
plt.ylabel('Average mAP')
plt.title('mAP Across Models Across mutiple rounds')
plt.savefig('mAP Across Models Across mutiple rounds.jpg')
plt.legend()
plt.show()
```
