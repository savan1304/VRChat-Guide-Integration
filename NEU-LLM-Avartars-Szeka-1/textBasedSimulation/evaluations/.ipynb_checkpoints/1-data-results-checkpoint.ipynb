```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Parameters (m, n)': ['(3, 3)', '(3, 5)', '(5, 3)', '(5, 5)', '(5, 7)', '(7, 5)', '(7, 7)'],
    'mAP Accross Models for Test Run 1': [2.785316971, 2.88120517, 2.341611295, 2.93594704, 2.769488107, 2.58801233, 2.589405872],
    'mAP Accross Models for Test Run 2': [2.831115471, 2.839172143, 2.767282138, 2.701529528, 2.816452691, 2.689142178, 2.649476326]
}

df = pd.DataFrame(data)

df['Average'] = df[['mAP Accross Models for Test Run 1', 'mAP Accross Models for Test Run 2']].mean(axis=1)

df = df.drop(columns=['mAP Accross Models for Test Run 1', 'mAP Accross Models for Test Run 2'])
df
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Parameters (m, n)'], df['Average'], marker='o', label='Average mAP')
plt.xlabel('Parameters (m, n)')
plt.ylabel('Average mAP')
plt.title('mAP Across Models Across mutiple rounds')
plt.savefig('mAP Across Models Across mutiple rounds.jpg')
plt.legend()
plt.show()
```

# scoring scenario

```python
import pandas as pd

data1 = {'1': [2.568169], '2': [2.982660], '3': [2.868403], '4': [2.780245], '5': [2.789377], '6': [2.649160], '7': [2.509239]}
df1 = pd.DataFrame(data1)

data2 = {'1': [2.338910935], '2': [3.366666667], '3': [2.922222222], '4': [3.012857143], '5': [2.412741841], '6': [2.566666667], '7': [2.522222222]}
data3 = {'1': [2.783506062], '2': [2.945094631], '3': [2.88028911], '4': [2.818639165], '5': [2.751984297], '6': [2.587557542], '7': [2.563544594]}
df2 = pd.DataFrame(data2)
df3 = pd.DataFrame(data3)
merged_df = pd.concat([df1, df2,df3], ignore_index=True)
merged_df.rename(columns=lambda x: x.replace('c', ''), inplace=True)
merged_df
```

```python
import pandas as pd
import matplotlib.pyplot as plt


x_values = merged_df.columns

y_values1 = merged_df.iloc[0].values
y_values2 = merged_df.iloc[1].values
y_values3 = merged_df.iloc[2].values

plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values1, label='Test 1 - GPT Response Set 1', marker='o')
plt.plot(x_values, y_values2, label='Test 2 - GPT Response Set 2', marker='o')
plt.plot(x_values, y_values3, label='Test 3 - GPT Response Set 2', marker='o')

plt.title('Senario Curves for Mutiple Test Runs')
plt.xlabel('Senarios')
plt.ylabel('Values')

plt.legend()
plt.savefig('avg-sen.jpg')
plt.show()
```
